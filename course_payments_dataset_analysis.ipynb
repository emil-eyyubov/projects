```python
# Группировка, Агрегация, Сортировка данных 
```

```python
all_money = df.money.sum()
round(all_money,2)
```

```python
# Посчитать суммарное кол-во денег по каждому курсу и городу и проранжировать их по убыванию
 # сгруппировали по городам, указав, что "title" - не индекс
                                                               # и посчитали сумму денег для каждого title
money_by_city_and_title = df.groupby(['title', 'city'], as_index=False) \
    .aggregate({'money':'sum'}) \
    .sort_values('money', ascending=False)  # лучше переносить методы для читабельности
money_by_city_and_title
```

```python
# Теперь отправляем обратно заказчику, сохраняем в сsv файл
# Пишем index=False, чтобы в файле не было индексов
money_by_city_and_title.to_csv('money_by_city_and_title.csv', index=False)
```

```python
# Задача 2. Нужно подсчитать для каждого курса сколько было суммарно денег и какое кол-во заказов было завершено
# Функция query (запрос) нужна для фильтрации значений
money_by_title = df.query("status == 'Завершен'") \
    .groupby('title', as_index=False) \
    .aggregate({'money':'sum', 'number': 'count'}) \
    .sort_values('money', ascending=False) \
    .rename(columns={'number': 'success_orders'}) # лучше переносить методы для читабельности
money_by_title
```

```python
# Автоматизация скучной работы. Например требуется каждый день отправлять отчет о каких то данных
```

```python
# Перед отправкой обработанных данных, нужно всегда делать самопроверку на логику
# Проверим, 6 ли курсов должно быть, совпадает ли суммарное кол-во денег
#print(df.title.unique())

today = datetime.today().strftime('%Y-%m-%d')
file_name = 'money_by_title_{}.csv'.format(today)

if int(df.money.sum()) == int(money_by_title.money.sum()):
    print('OK!!! FILE {} is written'.format(file_name))
    money_by_title.to_csv(file_name, index=False)
else:
    print('ERROR')
```
