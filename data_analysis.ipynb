```python
def read_n_agg(URL):
    df = pd.read_csv(URL, sep=';')
    return df.groupby('company').agg({'income':'mean'})
    
URL = 'https://stepik.org/media/attachments/lesson/359209/companies.csv'  
read_n_agg(URL)
```

```python
import pandas as pd

taxi = pd.read_csv('https://stepik.org/media/attachments/lesson/359240/taxi_peru.csv', sep=';', 
                parse_dates = ['start_at', 'end_at', 'arrived_at'])
taxi.head()
```

```python
taxi_sources = taxi.source.value_counts(normalize=True).round(decimals=2) * 100
taxi_sources
```

```python
import seaborn as sns
taxi_sources = taxi.source.value_counts(normalize=True).reset_index().rename(columns={'index':'source', 'source':'frequency'})
taxi_sources
```

```python
ax = sns.barplot(x='source', y='frequency', data=taxi_sources)
```

```python
ax = sns.countplot(x=taxi.icon)
```

```python
ax = sns.countplot(x='source', hue='end_state', data=taxi)
```

```python
driver_score_counts = taxi.driver_score.value_counts(normalize=True) \
             .mul(100) \
             .round(2) \
             .reset_index() \
             .rename(columns={'index':'driver_score','driver_score':'percentage'}) \
             .sort_values('driver_score')
driver_score_counts
```

```python
plot_of_drivers = sns.barplot(x='driver_score', y='percentage', data=driver_score_counts, color='blue', alpha=0.5)
plot_of_drivers.set(xlabel='Driver Score', ylabel='Percentage')
sns.despine()
```

```python
rider_score_counts = taxi.rider_score.value_counts(normalize=True) \
             .mul(100) \
             .round(2) \
             .reset_index() \
             .rename(columns={'index':'rider_score','rider_score':'percentage'}) \
             .sort_values('rider_score')
rider_score_counts
```

```python
plt.subplot(1,2,1)
plot_of_drivers = sns.barplot(x='driver_score', y='percentage', data=driver_score_counts, color='blue', alpha=0.5)
plot_of_drivers.set(xlabel='Driver Score', ylabel='Percentage')
sns.despine()
plt.subplot(1,2,2)
ax = sns.barplot(x='rider_score', y='percentage', data=rider_score_counts, color='blue', alpha=0.5)
ax.set(xlabel='Rider Score', ylabel='Percentage')
sns.despine()
```

```python
user_data = pd.read_csv('https://stepik.org/media/attachments/lesson/360348/user_data.csv')
user_data.head()
```

```python
logs = pd.read_csv('https://stepik.org/media/attachments/lesson/360348/logs.csv')
logs.head()
```

```python
logs.platform.nunique()
```

```python
logs_success = logs.query('success == True').groupby('client', as_index=False).agg({'success':'count'}).sort_values('success', ascending=False).head(10)
logs_success
```

```python
logs.query('success == True').platform.value_counts().idxmax()
```

```python
full_user_data = user_data.merge(logs, on='client')
full_user_data
```

```python
full_user_data.query('premium == True').platform.value_counts().idxmax()
```

```python
ax = sns.histplot(x='age', hue='premium', data=full_user_data, stat='probability', kde=True, binrange=(10,80))
```

```python
premium_users_age = full_user_data.query('premium == True').age
non_premium_users_age = full_user_data.query('premium == False').age
ax = sns.distplot(premium_users_age)
ax = sns.distplot(non_premium_users_age)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1) # figure, axis
dist_1 = sns.histplot(premium_users_age, kde=True, ax=ax[0], color='purple', binrange=(10,80))
dist_1 = sns.histplot(non_premium_users_age, kde=True, ax=ax[1], color='green', binrange=(10,80))
```

```python
# Постройте график распределения числа успешных операций: 

#x – число успешных операций
#y – число клиентов, совершивших такое количество успешных операций
client_success_data = full_user_data.query('success == True').groupby('client').agg({'success':'sum'})
client_success_data
```

```python
sns.histplot(client_success_data)
```

```python
client_success_data.success.value_counts()
```

```python
#изуализируйте число успешных операций, сделанных на платформе computer, в зависимости от возраста, используя sns.countplot. 
#По оси x – возраст, по оси y – число успешных операций.
#Если размер графика получается небольшим, можно добавить plt.figure(figsize=(12, 8)) в начало соответствующей ячейки. 

age_success = full_user_data.query('platform == "computer" & success == True') \
              .groupby('age', as_index=False) \
              .agg({'success':'count'}) \
              .rename(columns={'success':'count_of_successes'}) \
              .sort_values('count_of_successes', ascending=False)
age_success
```

```python
df = full_user_data.query('platform == "computer" & success == True')
plt.figure(figsize=(12, 8))
sns.countplot(x='age',data=df)
```
